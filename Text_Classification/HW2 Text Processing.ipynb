# 3. Identify the cleaning points

Text cleaning techniques
1. Tokenization and removing all punctuations
2. Normalize Text
3. Remove Stopwords
4. Perform Stemming and Lemmatization

# 4. Conduct text cleaning

## 4.1 Tokenization and removing all punctuations

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")
tokens = tokenizer.tokenize(raw)
print(tokens[0:100])
```

## 4.2 Normalize Text

Transform to lower cases

```python
tokens = [w.lower() for w in tokens]
print(tokens[0:100])
```

## 4.3 Remove Stopwords

```python
import nltk
stop_words = set(nltk.corpus.stopwords.words('english'))

tokens = [w for w in tokens if not w in stop_words]
print(tokens[0:100])
```

## 4.4 Perform Stemming and Lemmatization

Stemming

```python
porter = nltk.PorterStemmer()
tokens = [porter.stem(t) for t in tokens]
print(tokens[0:100])
```

Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()

tokens = [lemmatizer.lemmatize(l) for l in tokens]
print(tokens[0:100])
```
